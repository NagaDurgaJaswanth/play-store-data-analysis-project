```python
def clean_size(x):
    if x.isdigit():
        return float(x)
    elif(x[-1]=="M"):
        return float(x[:-1]) * 1e06
    elif(x[-1]=="k"):
        return float(x[:-1]) * 1e03
    else:
        return np.nan
```

```python
play["size_new"]=play["Size"].apply(clean_size)
```

```python
play["Size"]=play["size_new"]
play
```
